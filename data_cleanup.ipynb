```python
#read dataset and symptom severity csv
weights = pd.read_csv("Symptom-severity.csv")
dataset = pd.read_csv("dataset.csv")
weights = weights.drop(132)
weights
```

```python
#get symptom severity and associated symptoms as arrays
s = weights['Symptom'].iloc[:].values
severity = weights['weight'].iloc[:].values
```

```python
train = dataset.to_numpy() #dataset dataframe to array
train_clean = [ [0]*132 for i in range(4920)] #placeholder array for clean dataset
count = 0 #dataset row
#read each row in dataset
for i in train:
    #read each symptom in symptom list array
    for index,x in enumerate(symptoms): 
        #if symptom found in dataset row, set 1 in placeholder array
        if x in i:        
            train_clean[count][index]=1
    count=count+1
#reshape array
train_clean
tc = np.array(train_clean)
```
